```python
import time
```

```python
start_time=time.time()
current_time=start_time

fast_update = 1
slow_update = 5
```

```python
while True:
    print ("Main loop")
    
    if(int(current_time - start_time) % slow_update == 0):
        print ("5 Second loop")
               
    current_time=time.time()
    time.sleep(fast_update)
```
